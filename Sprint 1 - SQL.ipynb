# Ejercicio 1

El esquema creado tiene las siguientes características principales:

## Tablas y Variables:

### Tabla de Géneros (tb_genre)

Guarda información sobre los géneros de películas.
Variables destacadas: genre_id (primary key), genre_name (nombre del género).

### Tabla de Películas (tb_movie)

Almacena detalles de películas como título, fecha de lanzamiento y formato.
Variables destacadas: movie_id (primary key), movie_title (título de la película), movie_date (fecha de lanzamiento).

### Tabla de Roles (tb_role)

Contiene roles en la industria del cine, como actor o director.
Variables destacadas: role_id (primary key), role_name (nombre del rol).

### Tabla de Personas (tb_person)

Registra información sobre personas relacionadas con las películas.
Variables destacadas: person_id (primary keyID único), person_name (nombre de la persona), person_country (país de origen).

### Tabla de Relación Película-Persona (tb_movie_person)

Conecta películas, personas y roles, indicando si una película ganó un premio.
Variables destacadas: movie_id (ID de la película), person_id (ID de la persona), role_id (ID del rol), movie_award_ind (indicador de premio).

## Ejercicio 3

```python
SELECT tb_genre.genre_name, count(tb_movie.movie_id) AS Numero_Peliculas FROM tb_movie
        INNER JOIN tb_genre ON tb_movie.movie_genre_id=tb_genre.genre_id
        GROUP BY tb_genre.genre_id
        ORDER BY count(tb_movie.movie_id) DESC;
```

## Ejercicio 4

```python
SELECT tb_person.person_name, tb_movie_person.movie_id, count(tb_movie_person.role_id) AS max_roles 
      FROM tb_movie_person      
      LEFT JOIN tb_person ON tb_person.person_id = tb_movie_person.person_id     
      GROUP BY tb_person.person_name, tb_movie_person.movie_id     
```

```python
Image('SP01_T01_E04_A.png')
```

```python
SELECT tb_person.person_name, tb_movie_person.movie_id, count(tb_movie_person.role_id) AS max_roles     
      FROM tb_movie_person    
      LEFT JOIN tb_person ON tb_person.person_id = tb_movie_person.person_id   
      GROUP BY tb_movie_person.movie_id, tb_person.person_name   
      HAVING max_roles > 1 
```
